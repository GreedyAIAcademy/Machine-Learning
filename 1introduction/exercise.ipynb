```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./exercise/height.vs.temperature.csv')
```

```python
X = df['height'].values.reshape(-1,1)
Y = df['temperature'].values.reshape(-1,1)
```

```python
reg = LinearRegression()
reg.fit(X,Y)
```

```python
print('Regression formula: y = {:.5f}x + {:.2f}'.format(reg.coef_[0][0],reg.intercept_[0]))
```

```python
predict = reg.predict(X)
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(df['height'], df['temperature'], c ='black')
plt.plot(df['height'],predict)
plt.xlabel("height")
plt.ylabel("temperature")
plt.show()
```
